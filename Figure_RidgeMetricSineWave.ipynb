```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(1,1, figsize=(8, 3))
ax.set_ylim(-1, 2)

# number of samples
n = 1000

# Plot ridges
x = np.linspace(0, 6*np.pi, n)
y = ((np.cos(x+np.pi)+1)/2)
ax.plot(x, y, color='k', lw=3, zorder=3)

# # Plot zero crossing line
# ax.axhline(0.5, color='tab:orange', alpha=0.5)

# Plot square wave
y_square = y.copy()
y_square[y<0.5] = 0
y_square[y>0.5] = 1

ax.plot(x, y_square, color='grey', ls="--", zorder=1, alpha=0.7, label="Binary\nClassification")

# ANNOTATIONS

# Label Ridges
for i in range(3):
    text = f"$R_{i+1}$"
    x_loc = (i+0.5)*2*np.pi
    y_loc = 0
    ax.annotate(text=text, xy=(x_loc, y_loc), size=20, horizontalalignment='center')

## Ridge width
### set tick width and center
w_color = "darkturquoise"
w_w = 0.1
w_c = 1.4
w_xmin, w_xmax = np.array([0.5, 1.5])*np.pi

ax.vlines((w_xmin, w_xmax), w_c-w_w, w_c+w_w, color=w_color)
ax.hlines(w_c, w_xmin, w_xmax, color=w_color)
ax.annotate("w", xy=(1*np.pi, 1.5), color=w_color,  size=16, horizontalalignment='center')

text = r"$w_{R_1}=w$"
ax.annotate(text=text, xy=(1*np.pi, -0.5), size=20, color=w_color, horizontalalignment='center')


## Ridge Amplitude
a_color = "magenta"
a_xmin, a_xmax = np.array([2, 4])*np.pi

a_w = w_w*4

for i, loc in enumerate([a_xmin, a_xmax]):
    ax.vlines(loc, ymin=0, ymax=1, color=a_color, zorder=4)
    ax.hlines((1, -0.01), xmin=loc-a_w, xmax=loc+a_w, color=a_color, zorder=4)
    ax.annotate(f"$a_{i+1}$", xy=(loc, 1.1), color=a_color, size=16, horizontalalignment='center')
    
text = r"$a_{R_3}=\frac{a_1+a_2}{2}$"
ax.annotate(text=text, xy=(3*np.pi,-0.5), size=20, color=a_color, horizontalalignment='center')

## Ridge Spacing
s_color = "orangered"
s_xmin, s_xmax = np.array([3, 5])*np.pi

s_w = w_w
s_c = w_c

ax.vlines((s_xmin, s_xmax), s_c-s_w, s_c+s_w, color=s_color)
ax.hlines(s_c, s_xmin, s_xmax, color=s_color)
ax.annotate("s", xy=(4*np.pi, 1.5), color=s_color,  size=16, horizontalalignment='center')

text = r"$s_{R_3}=s$"
ax.annotate(text=text, xy=(5*np.pi, -0.5), size=20, color=s_color, horizontalalignment='center')

# General Figure formatting
ax.set_ylim(-0.5, 1.6)
ax.set_xlim(0, 22)
ax.set_axis_off()
ax.legend(frameon=False, loc="center right")
plt.tight_layout()

plt.savefig("/Users/avan/FLUD/BrazosScrolls/figures/ExampleRidgeClassificationAndMetrics.png", dpi=300)
```
